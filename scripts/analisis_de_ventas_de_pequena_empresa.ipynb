```
base_de_datos=""
ventas_totales=0
producto_mas_vendido=""
cantidad_vendida=0
ventas_enero=0
ventas_febrero=0
ventas_marzo=0
ventas_abril=0
ventas_mayo=0
ventas_junio=0
ventas_julio=0
ventas_agosto=0
ventas_septiembre=0
ventas_octubre=0
ventas_noviembre=0
ventas_diciembre=0
ventas_mensuales=[]
fecha=""
mes=""
meses=["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]

import csv

# Ruta relativa desde scripts/ hacia datos/
with open("../datos/sales_sample_2024.csv", "r") as archivo:
    lector=csv.reader(archivo,delimiter=",")
    next(lector)

    for i in lector:
      ventas_totales+=int(i[2])

      if int(i[2])>cantidad_vendida:
        cantidad_vendida=int(i[2])
        producto_mas_vendido=i[0]

      fecha=i[1]
      mes=int(fecha[5:7])
      if mes==1:
        ventas_enero+=int(i[2])
      elif mes==2:
        ventas_febrero+=int(i[2])
      elif mes==3:
        ventas_marzo+=int(i[2])
      elif mes==4:
        ventas_abril+=int(i[2])
      elif mes==5:
        ventas_mayo+=int(i[2])
      elif mes==6:
        ventas_junio+=int(i[2])
      elif mes==7:
        ventas_julio+=int(i[2])
      elif mes==8:
        ventas_agosto+=int(i[2])
      elif mes==9:
        ventas_septiembre+=int(i[2])
      elif mes==10:
        ventas_octubre+=int(i[2])
      elif mes==11:
        ventas_noviembre+=int(i[2])
      elif mes==12:
        ventas_diciembre+=int(i[2])

ventas_mensuales=[ventas_enero,ventas_febrero,ventas_marzo,ventas_abril,ventas_mayo,ventas_junio,ventas_julio,ventas_agosto,ventas_septiembre,ventas_octubre,ventas_noviembre,ventas_diciembre]

print(f"Las ventas totales son {ventas_totales}")
print(f"El producto mas vendido es {producto_mas_vendido} con una cantidad de {cantidad_vendida}")
print(f"""Las ventas mensuales fueron:
    Enero: {ventas_enero}
    Febrero: {ventas_febrero}
    Marzo: {ventas_marzo}
    Abril: {ventas_abril}
    Mayo: {ventas_mayo}
    Junio: {ventas_junio}
    Julio: {ventas_julio}
    Agosto: {ventas_agosto}
    Septiembre: {ventas_septiembre}
    Octubre: {ventas_octubre}
    Noviembre: {ventas_noviembre}
    Diciembre: {ventas_diciembre}""")

# Grafico evolucion mensual de ventas
import matplotlib.pyplot as plt

plt.plot(meses, ventas_mensuales)
plt.title("Evolución mensual de ventas")
plt.xlabel("Mes")
plt.ylabel("Ventas")

# Guardar gráfico en /resultados
plt.savefig("../resultados/evolucion_mensual_ventas.png")
plt.show()

```
