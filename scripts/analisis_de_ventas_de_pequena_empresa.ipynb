# 📊 Análisis de Ventas de una Pequeña Empresa
### Trabajo Práctico - Organización Empresarial (UTN)
Este script realiza la carga del dataset comercial, procesa las métricas de negocio principales e ilustra la evolución mensual de las ventas.

```
import csv
import matplotlib.pyplot as plt

# Inicialización de variables
ventas_totales = 0
producto_mas_vendido = ""
cantidad_vendida = 0
ventas_mensuales_dict = {i: 0 for i in range(1, 13)}
meses_nombres = ["Ene", "Feb", "Mar", "Abr", "May", "Jun", "Jul", "Ago", "Sep", "Oct", "Nov", "Dic"]

# 1. Carga de datos mediante ruta relativa funcional
with open("../datos/sales_sample_2024.csv", "r") as archivo:
    lector = csv.reader(archivo, delimiter=",")
    next(lector)  # Omitir cabecera

    for fila in lector:
        cantidad = int(fila[2])
        producto_id = fila[0]
        fecha = fila[1]
        
        # Acumular ventas totales
        ventas_totales += cantidad

        # Buscar producto más vendido
        if cantidad > cantidad_vendida:
            cantidad_vendida = cantidad
            producto_mas_vendido = producto_id

        # Agrupar ventas por mes
        mes = int(fecha[5:7])
        ventas_mensuales_dict[mes] += cantidad

# Conversión a lista para graficar
ventas_mensuales = [ventas_mensuales_dict[mes] for mes in range(1, 13)]

# 2. Presentación de métricas en consola
print(f"Las ventas totales acumuladas son: {ventas_totales}")
print(f"El producto más vendido es el ID {producto_mas_vendido} con {cantidad_vendida} unidades.\n")
print("Evolución de ventas mensuales:")
for nombre, monto in zip(meses_nombres, ventas_mensuales):
    print(f"    {nombre}: {monto}")

# 3. Generación del gráfico de evolución comercial
plt.figure(figsize=(10, 5))
plt.plot(meses_nombres, ventas_mensuales, marker='o', color='b', linestyle='-')
plt.title("Evolución Mensual de Ventas - Ejercicio Comercial 2024")
plt.xlabel("Meses del Año")
plt.ylabel("Volumen de Ventas")
plt.grid(True, linestyle='--', alpha=0.6)

# 4. Almacenamiento automático del resultado en la carpeta correspondiente
plt.savefig("../resultados/evolucion_mensual_ventas.png", dpi=300, bbox_inches='tight')
plt.show()
```
